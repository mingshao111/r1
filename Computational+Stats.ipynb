```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(facecolor=(1,0,0,.1))
plt.show()
```

```python
fig=plt.figure(figsize=plt.figaspect(2.0),facecolor=(1,0,0,.1))
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.set(xlim=[0.5,4.5], ylim=[-2,8], title='An Example Axes', ylabel='Y-Axis', xlabel='X-Axis')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.set_xlim([0.5,4.5])
ax.set_ylim([-2,8])
ax.set_title('A different Example')
ax.set_xlabel('X-axis-changed')
ax.set_ylabel('Y-axis-changed')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot([1,2,3,4],[10,20,30,40],color='lightblue',linewidth=3)
ax.scatter([0.3,3.8,1.2,2.5],[11,25,9,26],c=[1,3,2,4])
ax.set_xlim([0,4.5])
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].set(title='UL')
axes[0,1].set(title='UR')
axes[1,0].set(title='LL')
axes[1,1].set(title='LR')

for ax in axes.flat:
    ax.set(xticks=[], yticks=[])
    
plt.show()    
    
```

```python
x=np.linspace(0,10,100)
y1, y2, y3=np.cos(x), np.cos(x+1), np.cos(x+2)
names=['Signal1','Signal2','Signal3']

fig,axes=plt.subplots(nrows=3)

for ax,y,name in zip(axes, [y1,y2,y3], names):
    ax.plot(x,y,color='black')
    ax.set(xticks=[], yticks=[], title=name)
plt.show()

list(zip(axes, [y1,y2,y3], names))
```
